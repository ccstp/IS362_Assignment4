## Question 1

### What is the northernmost airport in the United States?

```python
# Sort airports table by latitude in descending order

airports.sort_values(['lat'], ascending=False)
```

The airports table has Dillant-Hopkins Airport as being at the highest latitude; however, it is listed as being in the Eastern time zone and a Google Maps search shows it to be in New Hampshire. 

With the northernmost state of the United States being Alaska, it was obvious that Dillant-Hopkins Airport could not be the northernmost airport. 

Based on this information, the airports table was filtered to only display airports in GMT -9, the time zone for Alaska.

```python
# Display only airports in GMT -9, the time zone for Alaska, which is the northernmost state of the United States

north = airports[(airports.tz == -9.0)]
north
```

The airports table was then sorted by latitude in descending order to determine the northernmost airport in the United States.

```python
# Sort airports table by latitude in descending order and return the northernmost airport in the United States

northernmost = north.sort_values(by='lat', ascending = False)
northernmost.head(1)
```

## Question 2

### What is the easternmost airport in the United States?

```python
# Sort airports table by longitude in descending order

airports.sort_values(['lon'], ascending=[False])
```

Since the question is asking for the easternmost airport in the United States, it was safe to assume that the airport would be located in GMT -5, the Eastern time zone.

The airports table shows Dillant-Hopkins Airport as being the easternmost airport in GMT -5; however, it is located in New Hampshire, which is not the easternmost state in the United States. It does not display a valid value in the tzone column.

Based on this information, the airports table was filtered to only display airports in GMT -5, the easternmost time zone in the United States, and to remove aiports with an invalid time zone.

```python
# Display only airports in GMT -5, which is the easternmost time zone in the United States

east = airports[(airports.tz == -5.0) & (airports.tzone).notnull()]
east
```

The airports table was then sorted by longitude in descending order to determine the easternmost airport in the United States.

```python
# Sort airports table by longitude in descending order and return the easternmost airport in the United States

easternmost = east.sort_values(by='lon', ascending = False)
easternmost.head(1)
```

## Question 3

### On February 12th, 2013, which New York area airport had the windiest weather?

```python
# Display weather data in the dataframe

weather.head()
```

```python
# Display weather data only for February 12, 2013 and sort by wind speed in descending order

weather[(weather.year == 2013) & (weather.month == 2) & (weather.day == 12)].sort_values('wind_speed', ascending = False)
```

The fastest wind speed ever recorded was 254mph ([ThoughtCo](https://www.thoughtco.com/fast-wind-speed-recorded-3444498)), so the wind speed of 1048.36058 is an outlier and likely incorrect.

As the wind speed was recorded hourly, it made sense to find the average wind speed in the 24-hour period for each NYC airport.

The outlier was removed and the recorded wind speeds were averaged to determine the average wind speed on February 12, 2013, for each of the NYC area airports.

```python
# Remove outlier and averaged the hourly recorded wind speed to determine average wind speed for the day for each NYC airport 

weather[(weather.year == 2013) & (weather.month == 2) & (weather.day == 12) & (weather.wind_speed < 350)]\
    .groupby('origin').wind_speed.mean()
```
